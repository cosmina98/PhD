```python
def hash_list(l):
    r=l[0]
    for e  in l:
        r=hash((r, e))
    return r
```

```python
import random
def permute_graph(G_orig):
    G = G_orig.copy()
    node_map = [u for u in G.nodes()]
    random.shuffle(node_map)
    node_map = {v:u for u,v in zip(node_map, G.nodes())}
    G = nx.relabel_nodes(G, node_map)
    return G
```

```python
def plot_graph(G):
    pos=nx.kamada_kawai_layout(G)
    node_labels = {u:G.nodes[u]['label'] for u in G.nodes()}
    nx.draw_networkx(G, pos, with_labels=False)
    nx.draw_networkx_labels(G,pos, labels=node_labels)
```

```python
import networkx as nx
n=8
m=4
G = nx.barabasi_albert_graph(n, m) 
label_alphabet = ['a','b','c']
for u in G.nodes():
    G.nodes[u]['label']=random.choice(label_alphabet)
nx.set_edge_attributes(G,1,'label')  
plot_graph(G)
```

```python
G2 = permute_graph(G)
plot_graph(G2)
```

```python
G3 = G.copy()
es = [e for e in G.edges()]
ns = [u for u in G.nodes()]
random.shuffle(ns)
G3.remove_edge(*es[0])
G3.add_edge(ns[0], ns[1], label=1)
plot_graph(G3)
```

```python
def hash_edge(G, u,v):
    return hash_list(sorted([hash(G.nodes[u]['label']),hash(G.nodes[v]['label']),hash(G.edges[u,v]['label'])]))

def hash_neighborhood(G,u):
    ns = list(G.neighbors(u))
    es = sorted([hash_edge(G,u,v) for v in ns])
    GH = hash_list(es)
    return GH

def graph_hash(G):
    es = sorted([hash_neighborhood(G,u) for u in sorted(G.nodes())])
    GH = hash_list(es)
    return GH
```

```python
graph_hash(G),graph_hash(G2), graph_hash(G3)
```
